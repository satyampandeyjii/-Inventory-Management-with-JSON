```python
# Importint JSON
import json

# Importing date & time
import time
datetime= time.ctime()


fd = open("Product record.json",'r')
js = fd.read()
fd.close()

# Creating product detail
record= json.loads(js)


# Creating Manu for customer
print("             MENU                    \n")
for key in record:
    print(key,record[key]['name'], record[key]['price'], record[key]['qun'])
    
print("_"*30)

# Now take customer input
user_name=str(input("Enter your name      : "))
user_mbl =str(input("Enter your mobile no : "))
user_mail= str(input("Enter your mail     : "))
user_in_id = str(input("Enter product ID  : "))
user_in_qun = int(input("Enter quantity   : "))

# Generating Bill--------------------------------

# checking condition
if(user_in_id in record):
    
    if (record[user_in_id]['qun'] ==0):
        print("-"*30)
        print("Sorry this product is not avilable at that time")
        print("-"*30)


    elif(record[user_in_id]['qun'] >= user_in_qun):
        print("\n-------------Bill-----------\n")
        print(user_name)
        print(user_mbl)
        print("-"*10)
        print("Name of product : ",record[user_in_id]['name'])
        print("Price           : Rs",record[user_in_id]['price'])
        print("Quantity        : ",user_in_qun)
        print("-"*30)
        total_bill = (record[user_in_id]['price'])*(user_in_qun)
        print("Billing Amount  : Rs ",total_bill)
        print("-"*30)


        # Update quantity of that product
        record[user_in_id]['qun'] -= user_in_qun

        # generating product sales detail
        sale = user_name + "," + user_mbl + "," + user_mail + "," + user_in_id + "," + record[user_in_id]['name'] + "," + str(record[user_in_id]['price']) + "," + str(user_in_qun) + "," + str(total_bill) + "," + datetime + "\n"

        # record it in Sales detail file
        fd=open("Sales detail.txt",'a')
        fd.write(sale)
        fd.close()


    else:
        print("_"*30)
        print(f"Sorry we have only ,{record[user_in_id]['qun']}  quantity of {record[user_in_id]['name']}")
        print("If you want them, press(Y/y) ")
        ch= str(input("Enter please: "))

        if(ch=='y' or ch=='Y'):

            print("\n-------------Bill-----------\n")
            print(user_name)
            print(user_mbl)
            print("-"*10)
            print("Name of product : ",record[user_in_id]['name'])
            print("Price           : Rs",record[user_in_id]['price'])
            print("Quantity        : ",record[user_in_id]['qun'])
            print("-"*30)
            total_bill = (record[user_in_id]['price'])*(record[user_in_id]['qun'])
            print("Billing Amount  : Rs ",total_bill)
            print("-"*30)

            # Update quantity of that product
            record[user_in_id]['qun'] -= record[user_in_id]['qun']

            # generate product sales detail 
            sale = user_name + "," + user_mbl + "," + user_mail + "," + str(user_in_id) + "," + record[user_in_id]['name'] + "," + str(record[user_in_id]['price']) + "," + str(record[user_in_id]['qun']) + "," + str(total_bill) + "," + datetime + "\n"   

            # record it in Sales detail file
            fd=open("Sales detail.txt",'a')
            fd.write(sale)
            fd.close()
            
else: 
    print("-"*30)
    print("Sorry we have not such product if this ID; ",user_in_id)
    print("please enter valid product ID" )
    print("-"*30)
    
        
print("Thank You Mr/Mrs ",user_name)
print("_"*30)


# saving & update all things in Product record.json file
js= json.dumps(record)
fd = open("Product record.json","w")
fd.write(js)
fd.close()
```
